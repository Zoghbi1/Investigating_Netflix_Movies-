```python
# Importing pandas and matplotlib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
# Start coding here! Use as many cells as you like 
netflix_df = netflix_df[np.logical_and(netflix_df["release_year"]>= 1990,netflix_df["release_year"]<=1999, netflix_df["type"] == "Movie")] 
netflix_df.head() 
```

```python
#visualize the data to view the distributions of durations for movies released in the 1990's 
plt.hist(netflix_df["duration"])
plt.xlabel("Movie Durations(mins)")
plt.ylabel("Number of Movies")
plt.title("Distribution of Movie's Durations")
plt.show() 
duration = 100
```

```python
#Count the number of short action movies from the 1990's
netflix_df = netflix_df[netflix_df["genre"] == "Action"]
num_act_mov = 0
for lab,row in netflix_df.iterrows():
    if row["duration"] < 90 :
        num_act_mov = num_act_mov + 1 
    else :
        num_act_mov = num_act_mov
short_movie_count = num_act_mov
print(short_movie_count)
```
